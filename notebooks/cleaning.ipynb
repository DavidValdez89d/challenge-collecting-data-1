```python
import pandas as pd
import ujson
import numpy as np
```

```python
def load_data():
    try:
        with open('../data/temp.json', 'r') as fd:
            print("Loading temp_data.")
            return ujson.load(fd)
    except:
        return dict()
        print("No dump found.")
```

```python
json = load_data()
```

```python
df = pd.DataFrame(json['existing_properties'])
df = df.set_index('url')
df.price = df.price.str.extract(r'.*[ \n](\d+) â‚¬').astype(float)
df['online'] = np.where(df['venue of the sale'] == "link to online sale", True, False)
df['living area'] = df['living area'].str.extract(r'(\d+)').astype(float)
df['plot surface'] = df['surface of the plot'].str.extract(r'(\d+)').astype(float)
df['kitchen type'] = df['kitchen type'].mask(df['kitchen type']=='installed', 1)
df['kitchen type'] = df['kitchen type'].mask(df['kitchen type']=='semi equipped', 2)
df['kitchen type'] = df['kitchen type'].mask(df['kitchen type']=='hyper equipped', 3)
df["terrace surface"] = df["terrace surface"].str.extract(r'(\d+)').astype(float)
df["garden surface"] = df["garden surface"].str.extract(r'(\d+)').astype(float)
df['furnished'] = df['furnished'].mask(df['furnished']=='no', 0)
df['furnished'] = df['furnished'].mask(df['furnished']=='yes', 1)
df['swimming pool'] = df['swimming pool'].mask(df['swimming pool']=='no', 0)
df['swimming pool'] = df['swimming pool'].mask(df['swimming pool']=='yes', 1)



clean_df = df[['city', 'price', 'bedrooms', 'online', 'living area', 'plot surface', 'kitchen type', 'terrace surface', 'furnished', 'garden surface', 'swimming pool', 'number of frontages', 'building condition', 'how many fireplaces?']]

for c in clean_df.columns:
    try:
        clean_df[c].loc[c, :] = clean_df[c].astype(float)
    except:
        pass
```

```python
clean_df.to_csv('../data/clean_properties.tsv', sep='\t')
```
