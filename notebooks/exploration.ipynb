```python
import numpy as np
from scipy import stats
import pandas as pd
import seaborn as sn
```

```python
df = pd.read_csv("../data/clean_properties.tsv", index_col='url', sep='\t')
df
```

```python
target = "price"
rows, col = df.shape

correlations_with_price = df.corrwith(df[target]).sort_values(ascending=False)
correlations = df.corr()
correlations = correlations.mask(np.tril(np.ones(correlations.shape)).astype(bool))
ranked_correlations = correlations.unstack().dropna().sort_values(ascending=False)
ranked_correlations = ranked_correlations[ranked_correlations < 1.]
most_correlated  = ranked_correlations[ranked_correlations.abs() > 2 * ranked_correlations.std()]
least_correlated = ranked_correlations[ranked_correlations.abs() < ranked_correlations.std()]

```
